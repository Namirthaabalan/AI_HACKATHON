```python
import pandas as pd
from googletrans import Translator

# Load the CSV file
df = pd.read_csv('/content/drive/MyDrive/AI_Hackathon/train.csv')

# Initialize the translator
translator = Translator()

# Function to translate text
def translate_text(text):
    try:
        translated = translator.translate(text, dest='en')
        return translated.text
    except Exception as e:
        return str(e)

# Apply translation to the 'crimeaditionalinfo' column
df['translated_info'] = df['crimeaditionalinfo'].apply(translate_text)

# Save the updated DataFrame to a new CSV file
df.to_csv('translated_output.csv', index=False)
```

```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Remove rows with NaN values in the 'crimeaditionalinfo' column
df = df.dropna(subset=['crimeaditionalinfo'])

# Text cleaning: ensure values are strings and remove non-alphanumeric characters
df['crimeaditionalinfo'] = df['crimeaditionalinfo'].apply(lambda x: re.sub(r'[^\w\s]', '', str(x)) if isinstance(x, str) else '')

# Convert to lowercase
df['crimeaditionalinfo'] = df['crimeaditionalinfo'].apply(lambda x: x.lower())

# Tokenization
df['tokens'] = df['crimeaditionalinfo'].apply(word_tokenize)

# Stopword removal
stop_words = set(stopwords.words('english'))
df['filtered_tokens'] = df['tokens'].apply(lambda tokens: [word for word in tokens if word not in stop_words])

# Stemming
porter = PorterStemmer()
df['stemmed_tokens'] = df['filtered_tokens'].apply(lambda tokens: [porter.stem(word) for word in tokens])

# Display the preprocessed DataFrame
print(df[['crimeaditionalinfo', 'stemmed_tokens']])
```

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.preprocessing import LabelEncoder
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import label_binarize
import numpy as np

# Step 1: Load the dataset
train_df = pd.read_csv('/content/drive/MyDrive/AI_Hackathon/train.csv')

# Step 2: Data Preprocessing for Training
# Convert labels to numerical values
label_encoder = LabelEncoder()
train_df['category'] = label_encoder.fit_transform(train_df['category'])

# Feature Extraction for Training
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = tfidf_vectorizer.fit_transform(train_df['crimeaditionalinfo'].fillna(''))
y_train = train_df['category']

# Step 3: Model Training
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

# Load the test dataset
test_df = pd.read_csv('/content/drive/MyDrive/AI_Hackathon/test.csv')

# Feature Extraction for Testing
X_test_tfidf = tfidf_vectorizer.transform(test_df['crimeaditionalinfo'].fillna(''))

# Step 4: Model Prediction
y_pred = model.predict(X_test_tfidf)

# Convert predictions back to original labels
y_pred_labels = label_encoder.inverse_transform(y_pred)

# Add predicted labels to the test dataset
test_df['predicted_category'] = y_pred_labels

# Model Evaluation on Training Data
y_train_pred = model.predict(X_train_tfidf)
accuracy = accuracy_score(y_train, y_train_pred)
print(f"Training Accuracy: {accuracy:.2f}")
print("Classification Report on Training Data:")
print(classification_report(y_train, y_train_pred, zero_division=1))
```

```python
test_df.to_csv('/content/drive/MyDrive/AI_Hackathon/Crimeinfo.csv', index=False)

# Print or visualize the results
print("Predictions saved to 'Crimeinfo.csv':")
print(test_df[['crimeaditionalinfo', 'predicted_category']])
```
